<h1 id=\"data_exploration\">数据处理</h1>


### 字符元素序列化

```python
Encode = ['action_type','combined_shot_type','shot_type','shot_zone_area','shot_zone_basic','shot_zone_range','matchup','opponent','period']
```

将第一列字符元素全部替换为数字，共55种

```python
for i in range(len(Encode)):
    list1 = df_train[Encode[i]].unique().tolist()
    list2 = list(range(len(list1)))
    dic = dict(zip(list1,list2))
    df_train[Encode[i]] = df_train[Encode[i]].replace(dic)
```

```python
df_train.head(5)
```

### 离散值独热编码

```python
df_train = pd.get_dummies(df_train, columns = Encode)
df_train.head(5)
```

### 标准化

```python
df_train_scaled = (df_train - df_train.min()) / (df_train.max() - df_train.min())
```

```python
df_train_scaled.head()
```

### 数据描述

```python
df_train_scaled.info()
```

```python
df_train_scaled.describe()
```

### 高相关过滤

```python
df_dt=df_train_scaled.drop('shot_made_flag', axis=1)
df_corr = df_dt.corr()
Titles = np.array(df_corr.columns).tolist()
df_dt.head()
```

```python
for i in range(len(Titles)):
    for j in range(i+1,len(Titles)):
        if(df_corr[Titles[i]][Titles[j]]>.6):      # 因为特征过多，为了更快降低维度，这里采用.6作为强相关，但是这种方法可能会忽略一些特征
            df_train_scaled.drop(Titles[i],axis=1)
```

```python
df_corr
```

```python
df_train_scaled.head()
```

可见，并不能删除任何一个值，因此我们不能通过这种方法降维

### 主成分分析

```python
X = df_train_scaled.drop(['shot_made_flag'], axis = 1)
y = df_train_scaled[['shot_made_flag']]
```

```python
rcParams['figure.figsize'] = 5,4

#Fitting the PCA algorithm with our Data
pca = PCA().fit(X)
plt.xlim(left = 50, right = 130)

#Plotting the Cumulative Summation of the Explained Variance
plt.plot(np.cumsum(pca.explained_variance_ratio_))

plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Pulsar Dataset Explained Variance')
plt.show()
```

由此图可以近似得到，当保留100个变量时，我们可以拟合99%的特征

```python
pca = PCA(n_components = 100)
X = pca.fit_transform(X)
len(X)
```

### 区分训练集和测试集

训练集80%，测试集20%

```python
msk = np.random.rand(len(df_train_scaled)) < 0.8
X_train = X[msk]
X_test = X[~msk]
y_train = y[msk]
y_test = y[~msk]
```

### 逻辑回归

```python
Linreg = LogisticRegression(penalty='l2').fit(X_train, y_train)
```

```python
pred_y = Linreg.predict(X_test)
acc_matrix = (pred_y == y_test.values.reshape(-1))
accuracy = np.sum(acc_matrix)/len(acc_matrix)
print('The accuracy is: %.2f%%' % (accuracy * 100))
```

### 决策树

```python
from sklearn.tree import DecisionTreeClassifier
DTclf = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
pred_y = DTclf.predict(X_test)
acc_matrix = (pred_y == y_test.values.reshape(-1))
accuracy = np.sum(acc_matrix)/len(acc_matrix)
print('The accuracy is: %.2f%%' % (accuracy * 100))
```

### 逻辑回归调参

```python
Z = list()
solver_all = ['newton-cg','lbfgs','liblinear','sag','saga']
for i in range(len(solver_all)):
    for C_extended in np.arange(0.01,0.06,0.001):
        Z.append(round(accuracy_score(y_test,LogisticRegression(solver=solver_all[i],C=C_extended).fit(X_train, y_train).predict(X_test)),3))
    plt.plot(np.arange(0.01,0.06,0.001),Z)
    Z = list()
```

可见，使用以上特征工程及调参，回归学习的最大准确度有近似70%（没能达到）

### 决策树调参

```python
Z = list()
for max_depth_r in np.arange(1,10):
    for j in np.arange(0.001,0.2,0.01):
        Z.append(round(accuracy_score(y_test,DecisionTreeClassifier(max_depth=max_depth_r,min_samples_split=j).fit(X_train, y_train).predict(X_test)),3))
    plt.plot(np.arange(0.001,0.2,0.01),Z)
    Z = list()
```

可见，使用以上特征工程及调参，决策树学习的最大准确度仍然为近似70%（没能达到）
